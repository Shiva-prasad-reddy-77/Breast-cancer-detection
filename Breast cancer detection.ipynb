Data cleaning

```python
df.drop(df.columns[[-1,0]],axis=1,inplace =True)
```

Feature Selection

```python
from sklearn.model_selection import train_test_split
```

```python
diag_map={'M':1,'B' :0}
```

```python
df['diagnosis']=df['diagnosis'].map(diag_map)
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python


# Assuming df is your original DataFrame and X is defined as you described
X = df[['radius_mean', 'perimeter_mean', 'area_mean', 'concavity_mean', 'concave points_mean', 'symmetry_mean']]

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the scaler to your data and transform it
X_scaled = scaler.fit_transform(X)

# Convert the scaled data back to a DataFrame with the same column names
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

X_scaled_df
```

```python
X_scaled_df
```

```python
y=df[['diagnosis']]
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_scaled_df,y,test_size=0.2,random_state=42)
```

```python
print(f"Shape of X_train : {X_train.shape}")
print(f"Shape of X_test  : {X_test.shape}")
```

```python
print(f"Shape of y_train : {y_train.shape}")
print(f"Shape of y_test  : {y_test.shape}")
```

Model -KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=7)
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,y_test)
```

model -logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr_y_pred=lr.predict(X_test)
```

```python
accuracy_score(lr_y_pred,y_test)
```

Model-Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb=GaussianNB(var_smoothing=1e-2)
```

```python
gnb.fit(X_train,y_train)
```

```python
gnb_y_pred=gnb.predict(X_test)
```

```python
accuracy_score(gnb_y_pred,y_test)
```
